```python
x = ["Alekhya","mounika","niharika"]
y = [89,98,76]
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a,b,str(b))
plt.show()
```

```python
x = ["Alekhya","mounika","niharika"]
y = [89,98,76]
plt.barh(x,y)
for a,b in zip(y,x):
    plt.text(a,b,str(a))
plt.show()
```

#### Histogram plot

- Histogram is an accurate grapichal representation of the distribution of numerical data.
- It estimates the probability distribution of continious values and it is introduced by karl Perason
- **bin** - the range of values
- Histogram is a 2d array visuvalization

- plt.hist()

```python
x = np.random.randn(1000)
plt.hist(x,histtype="step")
plt.show()
```

```python
help(plt.hist)
```

```python
x = np.random.randn(1000)
plt.hist(x,bins=20)
plt.show()
```

#### Box plot

- This is also know as whisker plot
- it is used to find out outerliers

- **plt.boxplot()**

-![](https://www.simplypsychology.org/boxplot.jpg)

```python
x = np.random.randint(10,100,101)
plt.boxplot(x)
plt.show()
```

```python
data = np.random.normal(100,20,200)
plt.boxplot(data)
plt.show()
```

```python
a = np.random.randint(50,100,10)
b = np.random.randint(40,90,10)
c = np.random.randint(34,109,10)
plt.boxplot([a,b,c],labels=["a","b","c"])
plt.show()
```

#### Pie chart

 - plt.pie()

```python
labels = ["Python","Data Analysis","Machine Learning","AI"]
val = [56,67,89,100]
colors = ["gold","magenta","pink","skyblue"]
plt.pie(val,labels=labels,colors = colors,autopct = "%1.2f%%")
plt.axis("equal")
plt.show()
```

```python
help(plt.pie)
```

```python
labels = ["Python","Data Analysis","Machine Learning","AI"]
val = [56,67,89,100]
colors = ["gold","magenta","pink","skyblue"]
e = (0,2,2,0)
plt.pie(val,labels=labels,colors = colors,autopct = "%1.2f%%",explode=e)
plt.axis("equal")
plt.show()
```

#### Saving the figure

```python
labels = ["Python","Data Analysis","Machine Learning","AI"]
val = [56,67,89,100]
colors = ["gold","magenta","pink","skyblue"]
e = (0,2,2,0)
plt.pie(val,labels=labels,colors = colors,autopct = "%1.2f%%",explode=e)
plt.axis("equal")
plt.savefig("pieplot.jpg") #d://folder name//filename.png
plt.show()
```
