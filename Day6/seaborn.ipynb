##### categorical distribution plots


- Box plot

- bar plot

```python
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
sns.set_context("paper",font_scale=2),#poster,paper,notebook,talk
plt.figure(figsize=(10,9))
sns.catplot(x = "species",y = "petal_length",data = df,kind = "bar")
```

```python
sns.set_style("white")
sns.set_context("notebook",font_scale=1),#poster,paper,notebook,talk
plt.figure(figsize=(10,9))
sns.catplot(x = "species",y = "sepal_width",data = df,kind = "bar")
```

##### Violin plot
- it is the combination of box plot as well as kde(kernal density estimation) plot

```python
sns.catplot(x = "species",y = "petal_width" ,data = df,kind = "violin")
```

```python
sns.catplot(x = "species",y = "sepal_length" ,data = df,kind = "violin")
```

#### Joint plot

- it combines multiple kinds of plots for getting the insites from the data

```python
sns.jointplot(x ="sepal_length",y = "sepal_width",data =  df)
```

```python
sns.jointplot(x = "sepal_length",y = "sepal_width",kind = "hex",data = df)
```

```python
sns.jointplot(x = "sepal_length",y = "sepal_width",kind = "reg",data = df)
```

```python
sns.regplot(x = "sepal_length",y = "sepal_width",data =df)
```

###### Correlation

- correlation is used to describe the linear relationship between two continious variables

```python
c  = df.corr()
```

```python
c
```

#### Heat Map

- Heat map is specialized chart that uses to represent data values in format of table.
- it mostly useful for when we need to plot a large data and complex data.


```python
sns.heatmap(c)
```

```python
sns.heatmap(c,annot= True)
```

```python
sns.heatmap(c,annot = True,cmap = "plasma",linecolor = "black",linewidth = 5)
```
